```python
import pandas as pd

df = pd.read_csv('prueba2.csv', sep=';', encoding='utf-8')
df
```

```python
# listame la columna genre_name de la tabla df sin repetir valores
df['genre_name'].unique()
```

```python

#Eliminando valores nulos
df = df.dropna()
df.shape

# Selecciona solo las columnas relevantes para el análisis de market basket
data_mba = df[['id_invoice', 'genre_name']]

# Elimina duplicados para tener solo una entrada por pedido y álbum
data_mba = data_mba.drop_duplicates()

# Crea una tabla pivote donde cada fila es un pedido, cada columna es un álbum y los valores son 1 si el género fue comprado en ese pedido, 0 de lo contrario
basket = pd.pivot_table(data_mba, index='id_invoice', columns='genre_name', aggfunc=lambda x: 1, fill_value=0)

basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_encode_plus = basket.applymap(encode_units)
basket_encode_plus
```

```python
basket_filter_plus = basket_encode_plus[(basket_encode_plus > 0).sum(axis=1) >= 2]
basket_filter_plus
```

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets_plus = apriori(basket_filter_plus, min_support=0.03,
                                 use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)

frequent_itemsets_plus['length'] = frequent_itemsets_plus['itemsets'].apply(lambda x: len(x))

frequent_itemsets_plus
```

```python
frequent_itemsets_plus[ (frequent_itemsets_plus['length']  == 2) &
                        (frequent_itemsets_plus['support'] >= 0.03) ]
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(frequent_itemsets_plus, metric='lift',
                  min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True).head(10).sort_values('lift', ascending=False)
```

A partir de los resultados de las reglas de asociación anteriores, se puede ver que 'World' y 'Rock, Latin, Metal' son los elementos que tienen mayor asociación entre sí, ya que estos dos elementos tienen el valor de "elevación" mas alto. A mayor grado de elevación, mayor será la asociación entre los elementos. Al contar con un valor de elevación 6.06, siendo superior a 1, significa que los elementos tienen una asociación alta. Por lo que estos géneros son muy buenos para venderse juntos.


Adicionalmente, se puede ver que el valor de soporte de (World) y (Rock, Latin, Metal) es de 0.777778, lo que quiere decir que hay un 77% del total de transacciones en las que estos articulos se vendieron juntos

De la confianza se puede extraer que un cliente tenderá acomprar Heavy Metal antes de comprar Rock o Metal

```python
rules = association_rules(frequent_itemsets_plus, metric='lift', min_threshold=1)
rules['antecedents_len'] = rules['antecedents'].apply(lambda x: len(x))
rules['consequents_len'] = rules['consequents'].apply(lambda x: len(x))

# Filtrar las reglas donde la longitud de antecedentes y consecuentes es 1
filtered_rules = rules[(rules['antecedents_len'] == 1) & (rules['consequents_len'] == 1)]

# Ordenar y mostrar las primeras 10 reglas
filtered_rules = filtered_rules.sort_values('lift', ascending=False).reset_index(drop=True).head(10)
filtered_rules
```

A partir de los resultados obtenidos de las reglas de asociación, se puede observar que los géneros musicales Metal y Heavy Metal presentan una fuerte correlación entre sí. Esta asociación se evidencia con un valor de "elevación" superior a los demás, específicamente un valor de 2.83, lo cual es considerablemente alto. Esto sugiere que estos dos géneros tienden a ser adquiridos conjuntamente, lo que los convierte en una excelente combinación para promociones o ventas cruzadas.

En cuanto a los géneros Latin y Soundtrack, el análisis revela un valor de soporte de 0.057. Esto significa que en el 5.7% del total de las transacciones, estos dos géneros se vendieron juntos. Aunque este porcentaje puede parecer pequeño, no debe ser ignorado ya que representa una oportunidad de venta conjunta.

El análisis de la confianza nos permite extraer información adicional interpretando el antecedente y el consecuente. En el caso de Heavy Metal y Metal, con un valor de confianza de 1, se puede inferir que un cliente que compra Heavy Metal probablemente también comprará Metal. Esta información es valiosa para la estrategia de marketing, ya que se podría ofrecer descuentos en el género Metal después de que un cliente haya comprado Heavy Metal, incentivando así la compra de ambos géneros.
